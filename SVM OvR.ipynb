```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
```

```python
data_train = pd.read_excel(your path) #change with training data path
```

```python
X_train = data_train["preprocess"].tolist()
y_train = data_train["label"].tolist()
```

```python
vectorizer = TfidfVectorizer()
train_vect = vectorizer.fit_transform(X_train)
```

```python
svc = SVC()
o_vs_r = OneVsRestClassifier(svc)
o_vs_r.fit(train_vect, y_train)
```
